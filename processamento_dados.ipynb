```python
#Records setter
import random

num_registros = 100
dados = [random.randint(1, 100) for _ in range(num_registros)]
```

```python
#Normal
import time

tempo_inicial = time.time()
soma = 0



for valor in dados:
  time.sleep(0.1)
  soma += valor

print(soma)
print(f"Demorou: {time.time() - tempo_inicial}")
```

```python
#Jobs
from joblib import Parallel, delayed

tempo_inicial = time.time()

def somar_valor(valor):
  global soma_total
  time.sleep(0.1)
  soma_total += valor

soma_total = 0
Parallel(n_jobs=4, require='sharedmem')(delayed(somar_valor)(valor) for valor in dados)

print(soma_total)
print(f"Demorou: {time.time() - tempo_inicial}")
```

```python
#Threads
import threading

tempo_inicial = time.time()

class CalculaThread(threading.Thread):
  def __init__(self, valor):
    super().__init__()
    self.valor = valor

  def run(self):
    global valor_total
    time.sleep(0.1)
    valor_total += self.valor

valor_total = 0
threads = [CalculaThread(valor) for valor in dados]

for thread in threads:
  thread.start()

for thread in threads:
  thread.join()

print(valor_total)
print(f"Demorou: {time.time() - tempo_inicial}")
```
